```python
def binary_to_decimal(binary: str) -> int:
    """
    function to convert binary to decimal.
    args:
        (binary, str) : input binary.
    return:
        int: given decimal number.
    raise:
        TypeError: if input is not str
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(binary, str):
        raise TypeError("Input must be a string")
    # validate that not input
    if not binary:
        raise ValueError("Input cannot be empty")
    # Binary validation
    if any(digit not in "01" for digit in binary):
        raise ValueError("Invalid binary number")

    # logic
    total = 0
    power = 0
    
    for digit in reversed(binary):
        if digit == "1":
            total = total + 2**power
        power = power + 1
    return total

# -- main execution --
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter binary number :")
        # calling the function
        result = binary_to_decimal(user_input)
        print(result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
