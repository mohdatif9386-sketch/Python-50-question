```python
def is_Palindrome(text: str) -> bool:
    """
    function to check whether a given string is palindrome
    Args:
        (text: str) - To check string is palindrome
    return: 
        bool: True- if string is a palindrome otherwise False
    raise:
        TypeError: if input not a string.
        ValueError: if input string is empty.
    """
    # validate that input type
    if not isinstance(text, str):
        raise TypeError("Input must be a string")
    # string must not be empty
    if text.strip() == "":
        raise ValueError("Input string can not be empty")
    else:
        text = text.strip().lower()
    # reverse a string
    reverse_str = ""
    for i in text:
        reverse_str = i + reverse_str
    return text == reverse_str

# --main execution--
if __name__=="__main__":
    try:
        # taking input from the user
        user_input = input("Enter your string to check is palindrome or not : ").strip().lower()
        # calling the function
        if is_Palindrome(user_input):
            print(f"{user_input} is a palindrome")
        else:
            print(f"{user_input} is not a palindrome")
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
