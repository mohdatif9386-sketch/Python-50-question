```python
from typing import List,Any
def remove_duplicate_item(List: List[Any]) -> List[Any]:
    """
    function to remove duplicate items in the original list using for loop.
    Args:
        (number: list[int]) : remove duplicate item.
    return:
        int: given a list withoout duplicate items.
    raise:
        TypeError: if input is empty
    """
    # validate that input type
    if not isinstance(List, list):
        raise TypeError("if input is not a list")
    # validate list is not empty
    if len(List) == 0:
        raise ValueError("if input must not be empty")
    # validate list has only one element
    if len(List) == 1:
        print("List has already one element. No duplicate to remove")
        return List
    # logic
    unique_list = []
    for i in List:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

# ---main execution---
if __name__ == "__main__":
    # taking list from the user 
    #number_list = [1, 2, 3, 4, 5, 6, 7, 3, 5, 1, 8, 9, 10, 7]
    # number_list = [12]
    number_list = ["atif", "ifrazi", "ifrazi", "atif", "khan", "dsh"]
    try:
        # calling the function
        result = remove_duplicate_item(number_list)
    except (TypeError, ValueError) as error:
        print("Error:", error)
    except Exception as error:
        print("something went wrong :",error)
    else:
        print("After removing duplicate : ", result)
```
