```python
def binary_serach(List: list, target: int) -> int:
    """
    function to Perform binary search on a sorted list.

    Args:
        arr (list): Sorted list of integers
        target (int): Element to search

    Returns:
        int: Index of target if found, otherwise -1

    Raises:
        TypeError: If input types are invalid
        ValueError: If list is empty
    """
    # validate that input type
    if not isinstance(List, list):
        raise TypeError("Input must be a list")
    # if input is empty
    if not List:
        raise ValueError("Input can not be empty")
    # validate that target type
    if not isinstance(target, int):
        raise TypeError("input target must be an integer")

    # logic
    left = 0
    right = len(List) -1

    while left <= right:
        mid = (left + right) //2

        if List[mid] == target:
            return mid
        elif List[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
# --- main execution ---
if __name__ == "__main__":
    # taking input from the user
    user_input = [10, 20, 30, 40, 50, 60, 70, 80]
    target_input = 50

    try:
        # calling the function
        result = binary_serach(user_input, target_input)
        if result != -1:
            print(f"Element found at index : {result}")
        else:
            print("Element is not found in the list")
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)
```
