```python
def sort_dict_value(data):
    """
    fucntion to sort dictionary using by key.
    raise:
        TypeError: if input is not a dict.
    """
    if not isinstance(data, dict):
        raise TypeError("input must be a dict")
    # logic
    return sorted(data.items(), key= lambda x: x[1], reverse= False)

# -- main execution --
if __name__ == "__main__":
    try:
        user_input = {"banana" : 5, "apple": 2, "cherry": 8, "date": 3}
        # calling the function
        result = sort_dict_value(user_input)
        # print and display the result
        print(result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
