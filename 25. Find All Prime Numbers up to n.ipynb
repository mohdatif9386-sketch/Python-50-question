```python
def all_prime_number(n: int) -> list:
    """
    function to find all prime numbers up to n.

    Args:
        n (int): Upper limit

    Returns:
        list: List of prime numbers up to n

    Raises:
        TypeError: if input is not a number
        ValueError: if input is empty
    """
    # validate that input type
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    
    # if input is negative number
    if n < 0:
        raise ValueError("Input must be a positive number")

    # logic
    prime_number = []
    for num in range(2, n + 1):
        for i in range(2, int(num ** 0.5)+1):
            if num % i == 0:
                break
        else:
            prime_number.append(num)
    return prime_number

# --- main execution ---
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter the number :")
        # if input is empty
        if not user_input:
            raise ValueError("Input cannot be empty")
            
        # converting inside try block
        try:
            number = int(user_input)
        except ValueError:
            raise TypeError("Only integer input should be taken from the user.")
            
        # calling the function
        result = all_prime_number(number)
        print(f"All Prime Number List : {result}")
        
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)
```
