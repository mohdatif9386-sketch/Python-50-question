```python
def largest_number(input_number: list[float]) -> float:
    """
    function to find the largest element number using for loop.
    Args:
        input_number (input_numbers: list[float]): A list contaning numeric value (int or float)
    Return: 
        float: the largest element in the list
    raise:
        TypeError: if input is not list or containing non-numerical value
        ValueError: if the list is empty
    """
    # validate that input type in the numeric value
    if not isinstance(input_number, list):
        raise TypeError("input must be a list of number (int / float)")
    # validate that list is not empty
    if not input_number:
        raise ValueError("list can not be empty")
    # validate that element should be numeric
    for num in input_number:
        if not isinstance(num, (int,float)):
            raise TypeError("All elements must be numeric")
    largest = input_number[0]
    for num in input_number:
        if num > largest:
            largest = num
    return largest

# -- main execution-- 
if __name__=="__main__":
    try:
        # taking input from the user
        user_input = input("Enter your number seperated the space int or float only : ")
        #  converting user_input into number
        number_list = [int(i) for i in user_input.split()]
        # calling the function
        result = largest_number(number_list)
        # print and display the result
        print(f"Largest number is : {result}")
    except (TypeError, ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
```
