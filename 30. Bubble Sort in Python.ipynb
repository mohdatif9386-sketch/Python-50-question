```python
def bubble_sort(list1: list) -> list:
    """
    function to sort a list of integer in acsending order using bubbe sort algorithm.
    args:
        list1: list -- a list containing of integer.
    return:
        the sorted list in acsending order.
    raise:
        TypeError: if input is not a list.
        ValueError: Input cannot be empty.
    """
    # validate that input type
    if not isinstance(list1, list):
        raise TypeError("Input must be a list in integer")
    # validate that input is empty
    if not list1:
        raise ValueError("Input cannot be empty")

    # logic
    lenght = len(list1)
    for i in range(lenght):
        for j in range(0, lenght -i -1):
            if list1[j] > list1[j+1]:
                list1[j], list1[j+1] = list1[j+1], list1[j]
    return list1

# -- main execution --
if __name__ == "__main__":
    try:
        # taking input from the user
        
        # user_input = [10, 90, 50, 30, 20, 80, 70, 40, 60, 100]
        user_input = ["j", "h", "A", "d"]
        # user_input = []
        
        # calling the function
        result = bubble_sort(user_input)
        # print and display the result
        print("BUBBLE_SORT :", result)
    except Exception as e:
        print("Error :", e)
```
