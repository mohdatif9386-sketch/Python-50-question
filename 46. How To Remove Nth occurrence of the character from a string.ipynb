```python
def remove_nth_occurrence(string: str, char_to_remove: str, n: int) -> str:
    """
    function to remove Nth occurrence of the character from a string.
    args:
        (string: str, char_to_remove: str, n: int) - input there.
    return
        str: remove nth occurrnec.
    raise:
        TypeError: if input is not a strng
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(string, str) or not isinstance(char_to_remove, str):
        raise TypeError("Input must be a string")
        
    if not isinstance(n, int):
        raise TypeError("input n must be an integer")

    # Value validation
    if not string:
        raise ValueError("String cannot be empty")

    if not char_to_remove:
        raise ValueError("Character to remove cannot be empty")

    if len(char_to_remove) != 1:
        raise ValueError("Only one character allowed")

    if n <= 0:
        raise ValueError("n must be greater than 0")


    # logic
    occurrence_count = 0
    result_characters = []
    for current_character in string:
        if current_character == char_to_remove:
            occurrence_count += 1
            if occurrence_count == n:
                continue
        result_characters.append(current_character)
    return "".join(result_characters)

# -- main execution --
if __name__ == "__main__":
    try:
        # taking input from the user
        input_text = input("Enter string to remove nth occurrence :")
        remove_to_char = input("Enter to remove character :")
        count = int(input("Enter occurrence count :"))

        # calling the function
        result = remove_nth_occurrence(input_text, remove_to_char, count)
        # print and dispaly the result
        print("Remove nth occurrence :", result)
        
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
