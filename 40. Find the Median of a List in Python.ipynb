```python
def find_median(num_list: list) -> int:
    """
    function to find the median number of the list.
    args:
        (num_list: list) -- input number list
    return:
        int: median number of the list.
    raise:
        TypeError: if input is not a list.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(num_list, list):
        raise TypeError("Input must be a list")
    # validate that not input
    if not num_list:
        raise ValueError("Input cannot be empty")

    # logic
    num_list.sort()
    list_length = len(num_list)
    mid = list_length // 2

    if list_length % 2 != 0:
        return num_list[mid]
    else:
        return (num_list[mid-1] + num_list[mid]) / 2

# -- main execution --
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = [1, 2, 3, 4, 5, 6, 7, 8]
        # calling the function
        result = find_median(user_input)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
    else:
        print(f"Median number : {result}")
```
