```python
def reverse_word(sentence: str) -> str:
    """
    function to reverse the word in a given sentence.

    Args:
        (sentence: str) - Input sentence to reverse

    Return:
        str: Reversed sentence

    raise:
        TypeError: if input is not a string.
        ValueError: if input is empty
    """
    # validate that input type
    if not isinstance(sentence, str):
        raise TypeError("Input must be a string")
    # if input is empty
    if not sentence:
        raise ValueError("Input cannot be empty")

    # logic
    word = sentence.split()
    reverse = word[::-1]
    return " ".join(reverse)
    
# ---main execution---
if __name__ =="__main__":
    try:
        #taking input from the user
        user_input = input("Enter sentence to reverse the word : ")
        # calling the function
        result = reverse_word(user_input)
        #print and display the result
        print(f"Reversed word : {result}")
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)
```
