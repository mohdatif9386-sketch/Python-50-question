```python
def list_to_dict(list1: list) -> dict:
    """
    function to convert list of tuple to dictionary.
    args:
        (list1: list) - input list of tuple
    return:
        dict: Converts the given list into a dictionary.
    raise:
        TypeError: if input is empty.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(list1, list):
        raise TypeError("input must be a list")
    # validate that input is empty
    if len(list1) == 0:
        raise ValueError("Input cannot be empty")

    # logic
    dict1 = {}
    for key, value in list1:
        dict1[key] = value
    return dict1
# --- main execution --
if __name__ == "__main__":
    # user_input = [("A", 1), ("B", 2), ("C", 3), ("D", 4)]
    user_input = [("s", 2)]
    try:
        #CALLING THE FUCNTION
        result = list_to_dict(user_input)
        # print and display the result
        print(result)
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
