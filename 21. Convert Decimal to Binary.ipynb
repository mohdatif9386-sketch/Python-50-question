```python
def decimal_to_binary(number: int) -> str:
    """
    function to convert a decimal number into binary.

    Args:
        number (int): A non-negative decimal integer.

    Returns:
        str: Binary representation of the number.

    Raises:
        ValueError: If the number is negative.
        TypeError: If input is not an integer.
    """
    # valodate that input type
    if not isinstance(number, int):
        raise TypeError("Input must be an integer")
    # if number is negative
    if number < 0:
        raise ValueError("input must be a positive number")

    # logic
    binary = ""
    while number > 0:
        remainder = number % 2
        binary = str(remainder) + binary
        number //= 2
    return binary

# ---main execution---
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter the number : ")
        if not user_input:
            raise ValueError("input cannot be empty")
            
        #converting inside try block
        try:
            numbers = int(user_input)
        except ValueError:
            raise TypeError("Input must be a valid integer")
            
        # calling the function
        result = decimal_to_binary(numbers)
        
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)
    #print and display the result
    else:
        print("Binary", result)
```
