```python
def power_of_two(n: int) -> bool:
    """
    function yo calculae the power of two.
    args:
        n: int -- input number
    return:
        true: if the power of two otherwise False
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is empty
    """
    # validate that input type
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    # negative chack
    if n < 0:
        raise ValueError("Input must be a positive number")
    # logic
    if n > 0 and n & (n-1) == 0:
        return True, (n).bit_length() -1
    return False, None

# -- main exection --
if __name__ == "__main__":
    try:
        #taking input from the user
        user_input = input("Enter tye number to check of power :")
        # validate that input is empty
        if not user_input:
            raise ValueError("Input cannot be empty")
        # converting inside try block
        try:
            number = int(user_input)
        except ValueError:
            raise TypeError("Input must be a integer")
        #calling the function
        result = power_of_two(number)
        #print and display the result
        print("Power of two = ", result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
