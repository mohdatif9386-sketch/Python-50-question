```python
def generate_Subset(numbers: list[int]) -> list[list[int]]:
    """
    Generate all possible subsets (power set) of a list of numbers.

    Args:
        numbers (list[int]): Input list of integers.

    Returns:
        list[list[int]]: A list containing all subsets of the input list.

    Raises:
        TypeError: If input is not a list.

    Example:
        >>> generate_subsets([1, 2, 3])
        [[], [1], [2], [1, 2], [3], [1, 3], [2, 3], [1, 2, 3]]
    """
    # validate that input type
    if not isinstance(numbers, list):
        raise TypeError("Input must be a list")
    # validate that input is empty
    if not numbers:
        raise ValueError("Input cannot be empty")

    # logic
    subsets = [[]]
    for num in numbers:
        new_subsets = []
        for subset in subsets:
            new_subsets.append(subset + [num])
        subsets.extend(new_subsets)
    return subsets

# -- main execution --
if __name__ == "__main__":
    # taking input from the user
    user_input = [1, 2, 3, 4]
    try:
        #calling the function
        result = generate_Subset(user_input)
        print(result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
