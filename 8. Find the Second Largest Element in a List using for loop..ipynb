```python
from typing import List
def second_largest(number: list[int]) -> int:
    """
    find the second largest unique in a list without using sorting.
    Args:
        (number: list[int]) - a list of integer.
    returns:
        int: the second largest unique value.
    raise:
        TypeError: if input is not a list or elements are not integer.
        ValueError: if the list has fewer than 2 unique value
    """
    # validate that type input is a lits
    if not isinstance(number, list):
        raise TypeError("input must be a list of integer.")
    # validate that all elements in the list are integer
    if any(not isinstance(current_number, int) for current_number in number):
        raise TypeError("all element of list should be integer")
    # check list has atleast 2 elements
    if len(number) < 2:
        raise ValueError("list should contain atleast two arguments.")

    if number[0] > number[1]:
        largest = number[0]
        second = number[1]
    else:
        largest = number[1]
        second = number[0]
    
    for current_number in number[2:]:
        if current_number > largest:
            second = largest
            largest = current_number
        elif current_number != largest and current_number > second:
            second = current_number
    if second == largest:
        raise ValueError("list must contain atleast two unique element")
    return second

# ---main execution---
if __name__=="__main__":
    try:
        # taking number_list from the user
        number_list = [1, 2, 3, 99, 12, 55, 77, 88]
        # calling the function
        result = second_largest(number_list)
        # Exception Handling
    except (TypeError, ValueError) as error:
        print(f"Error: {error}")
    except Exception as e:
        print(e)
    else:
        print("second largest:", result)
```
