```python
def is_leap_year(year: int) -> bool:
    """
    function to check whether a given year is a leap year. 

    Args:
        year: int - Input year.

    Return:
        True - if year is leap otherwise False.
        
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is empty.
    """
    
    # validate that input type
    if not isinstance(year, int):
        raise TypeError("Input year must be an integer")
    # Input year is grater than 1582
    if year < 1582:
        raise ValueError("Input year must be after 1582.")
        
    #logic
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# --- main execution ---
if __name__ == "__main__":
    try:
        #taking input from the user
        user_input = input("Enter the year : ")
        # if input is empty
        if not user_input:
            raise ValueError("Input cannot be empty")

        try:
            # converting inside try block
            year = int(user_input)
        except ValueError:
            raise TypeError("Input must be an integer")
            
        # calling the function
        if is_leap_year(year):
            print(f"{year} is leap year")
        else:
            print(f"{year} is not a leap year")
                
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Unexpected Error", e )
```
