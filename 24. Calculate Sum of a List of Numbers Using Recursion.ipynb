```python
def sum_of_list(number: list) -> int:
    """
    function to calculate the sum of elements in a list using recursion.

    Args:
        numbers (list): List of integers

    Returns:
        int: Sum of all elements in the list

    Raises: 
        TypeError: if input is not a n integer
        ValueError: if input is empty
    """
    # validate that input type
    if not isinstance(number, list):
        raise TypeError("Input must be a list")
    # if input is empty
    if not number:
        raise ValueError("Input cannot be empty")
    
    # logic
    total = 0
    for i in number:
        total = total + i
    return total

# --- main execution ---
if __name__ == "__main__":
    #taking input from the user
    user_input = [1, 2, 3, 4, 5, 6, 7]
    try:
        # calling the function
        result = sum_of_list(user_input)

    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)

    else:
        print("Sum of list :", result)
        
```
