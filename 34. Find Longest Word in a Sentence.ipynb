```python
import string
def longest_word_finder(text: str) -> list:
    """
    function to find the longest word in the string.

    Args:
        text: str -- input string 

    Returns:
        list: given longest word

    Raises:
        TypeError: if input is not a string
        ValueError: if input is empty
    """
    # validate that input type
    if not isinstance(text, str):
        raise TypeError("Input must be a string")
    # validate that not input
    if not text:
        raise ValueError("Input cannot be empty")

    # logic
    text = text.lower()
    pattern = str.maketrans("", "", string.punctuation)
    clean_text = text.translate(pattern)
    words_list = clean_text.lower().split()
    
    word_count_dict = {}
    for word in words_list:
        if word not in word_count_dict:
            word_count_dict[word] = len(word)
            
    max_len = max(word_count_dict.values())
    
    longest_word = [key for key, value in word_count_dict.items() if value == max_len]
    return longest_word

# -- main execution--
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter string to find the longest word :")
        
        # calling the function
        result = longest_word_finder(user_input)
        # print and display the result
        print("Longest word :", result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
