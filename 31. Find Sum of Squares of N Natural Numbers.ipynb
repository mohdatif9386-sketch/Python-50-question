```python
def sum_of_square(n: int) -> int:
    """
    functiob to sum of square of n natural number.
    args:
        (n: int) -- input number
    return:
        the given number of sum square.
    raise:
        TypeError: if input is not an integer
        ValueError: if input is empty
    """ 
    # validate that input type
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")

    # logic
    result = n * (n+1) * (2*n + 1) // 6
    return result

# -- main execution --
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter the number :")
        
        # if input is empty
        if not user_input:
            raise ValueError("Input cannot be empty")
        # converting inside try block
        try:
            number = int(user_input)
        except ValueError:
            raise TypeError("input must be an integer")
            
        # calling the function
        result = sum_of_square(number)
        #print and display the result
        print(result)
        
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except (Exception) as e:
        print("Error :", e)
```
