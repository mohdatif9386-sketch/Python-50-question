```python
def fibonacci_sequence(number: int) -> int:
    """
    function to number convert fibonacci sequence using iterative approach.
    Args:
        number: int - input number 
    return:
        int: return fibonacci sequence.
    raises:
        TypeError: if input not an integer.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(number, int):
        raise TypeError("input must be an integer")
    # if input number is negative number
    if number < 0:
        raise ValueError("input must be a positive number")
    # logic
    a = 0 
    b = 1
    for i in range(number):
        print(a, end = " ")
        a, b = b, a+b
# --- main execution---
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter Number : ")
        # if input is empty
        if not number:
            raise ValueError("if input cannot be empty")
            
        # converting inside try block
        try:
            numbers = int(user_input)
        except ValueError:
            raise TypeError("Input must be an integer")
            
        # calling the function
        result = fibonacci_sequence(numbers)
        # print and display the result
        print("fibonacci sequence")
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
