```python
def find_missing_number(number: list[int]) -> int:
    """
    Function to find the missing number in a consecutive list.

    Args:
        number (list[int]): List of consecutive numbers with one missing value.

    Returns:
        int: The missing number.

    Raises:
        TypeError: if input is not a list.
        ValueError: if input is empty.
    
    Example:
        Input:  [1, 2, 3, 5, 6]
        Output: 4
    """
    # validate that input type.
    if not isinstance(number, list):
        raise TypeError("Input must be a list integer only")
    
    if not number:
        raise ValueError("Input cannot be empty")

    # logic
    for i in range(number[0], number[-1]):
        if i not in number:
            return i
    return find_missing_number(number)

# --- main execution ---
if __name__ == "__main__":
    user_input = [1, 2, 3, 5, 6]
    try:
        result = find_missing_number(user_input)
        print(result)
        
    except FileNotFoundError:
        print("File not found")

    except ValueError as e:
        print("Value Error:", e)

    except Exception as e:
        print("Unexpected Error:", e)
```
