```python
list1 = [[1, 2, 333], [4, 5, 666], [7, 8, 999]]
flatten_list = []
for i in list1:
    for y in i:
        flatten_list.append(y)
print(flatten_list)
```

```python
def flatten_list(nested_list: list) -> list:
    """
    function to convert nested list into flatten list.
    args:
        (nested_list: list) -- create a nested list
    return:
        flatten list
    raise:
        TypeError: iF INPUT is not a list
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(nested_list, list):
        raise TypeError("Input must be a nested list")
    # validate that input is empty
    if not nested_list:
        raise ValueError("Input cannot be empty")
    # logic
    list1 = []
    for i in nested_list:
        for item in i:
            list1.append(item)
    return list1
# ---main execution---
if __name__ == "__main__":
    try:
        #taking input from the user
        user_input = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
        # user_input = []
        # calling the function
        result = flatten_list(user_input)
        # print and display the result
        print("Flatten list :", result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
