```python
def is_anagram(text1: str, text2: str) -> bool:
    """
    function to check whether two string are anagram using character counting.
    space are ingnored and comparison is case-insensitive.

    Args:
        text1: str, text2: str -- first input and second input.
        
    return:
        bool: True - if text are anagram and False otherwise.

    raises:
        TypeError: If input is not a string.
    """
    # validate that input type
    if not isinstance(text1, str) or not isinstance(text2, str):
        raise TypeError("Input must be a string")
        
    text1 = text1.replace(" ", "").lower()
    text2 = text2.replace(" ", "").lower()

    #if input is empty
    if not text1 or not text2:
        raise ValueError("input cannot be empty")

    # if len is not equal
    if len(text1) != len(text2):
        return False
        
    # logic
    char_count = {}
    for char in text1.lower():
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
   
    for char in text2.lower():
        if char not in char_count:
            return False
        char_count[char] -= 1
        if char_count[char] < 0:
            return False
    else:
        return True

# ---main execution---
if __name__ == "__main__":
    try:
        # taking inout from the user
        user_input1 = input("Enter string 1 : ")
        user_input2 = input("Enter string 2 : ")
        
        # calling the function
        result = is_anagram(user_input1, user_input2)
        if result:
            print("your string is anagram")
        else:
            print("your string is not a anagram")
            
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
