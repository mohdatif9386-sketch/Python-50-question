```python
def sum_of_digits(n: int) -> int:
    """
    calculate the sum of digits number of a non-negative integer.
    Parameter:
        n: (int) -- The number whose didgt will be summed.
    return:
        int: sum of the digit.
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is negative.
    """
    # validate that input type
    if not isinstance(n, int):
        raise TypeError("input must be a integer")
    if n < 0:
        raise ValueError("input must be a positive number")
    total = 0
    while n > 0:
        total = total + (n % 10)
        n = n // 10
    return total

# -- main execution --
if __name__=="__main__":
    
    # taking input from the user
    number = input("Enter number to calculate the sum of digit : ")
    try:
        num = int(number)
    except ValueError:
        print("Error:", "please enter valid integer only.")
        # calling the fumction
    else:
        try:
            result = sum_of_digits(num)
            # print and dispaly the result
            print(f"Sum of digit is : {result}")
        except (TypeError, ValueError) as e:
            print("Error",e)
        except Exception as e:
            print(e)
```
