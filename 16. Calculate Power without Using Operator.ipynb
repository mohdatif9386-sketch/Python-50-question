```python
def calculate_power(base_value: int, power_value: int) -> int:
    """
    funcion to Calculate Power without Using Operator.

    Args: 
        number: int, power_value: int -- Input number and power value.

    Return:
        int: Number of given power.
        
     Raise:
         TypeError: if input is not an integer.
         ValueError: if input is empty.
    """

    # validate that input type
    if not isinstance(base_value, int) or not isinstance(power_value, int):
        raise TypeError("both input must be a integer")

    # logic
    result = 1
    for i in range(power_value):
        result = result * base_value
    return result
    
# ---main execution---
if __name__ == "__main__":
    try:
        # taking input from the user
        base_number = input("Enter the base number :")
        power_value = input("Enter the power value : ")

        # if input is empty
        if not base_number or not power_value:
            raise TypeError("Input cannot be empty")

        try:
            # converting inside try bloack
            base_number = int(base_number)
            power_value = int(power_value)
        except ValueError:
            raise TypeError("base number and power number must be a valid integer")
    
        # calling the function
        result = calculate_power(base_number, power_value)
        print(f"Power of the number is: 8: {result}")
        
    except (TypeError, ValueError) as error:
        print("Error :", error)
    except Exception as error:
        print("Unexpected Error :", error)
```
