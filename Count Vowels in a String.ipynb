```python
# Count Vowels in a String
```

```python
name = "shehzadi"
vowels = "aeiou"
count = 0
for i in name.lower():
    if i in vowels:
        count = count + 1
print(count)
```

```python
def count_vowels(text: str) -> int:
    """
    function to count the number of vowels in the string.
    Args:
        (text: str) - The input string.
    return:
        int: number of vowels in the string.
    raises:
        TypeError: if input is not a string. 
    """
    # validate that input type
    if not isinstance(text, str):
        raise TypeError("input must be a string")
    if not text:
        raise ValueError("Input string can not be empty")
        
    # Logic
    vowels = "aeiou"
    count = 0
    for i in text.lower():
        if i in vowels:
            count = count + 1
    return count

# --main execuution--
if __name__=="__main__":
    try:
        # Takig input from the user
        user_input = input("Enter text to count the vowels : ").lower()
        # calling the function
        result = count_vowels(user_input)
        # print and display the result
        print(f"Total number of vowels in {user_input} is : {result}")
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
