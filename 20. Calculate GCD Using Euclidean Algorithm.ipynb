```python
def calculate_GCD(num1: int, num2: int) -> int:
    """
    function to calculate GCD of two numbers using Euclidean Algorithm.

    Args:
        num1 (int): first number
        num2 (int): second number

    Returns:
        int: GCD of num1 and num2

    Raises:
        TypeError: if input is not an integer.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(num1, int) or not isinstance(num2, int):
        raise TypeError("Both input must be an integer")
    if not num1 or not num2:
        raise ValueError("Input can not be empty")
    #logic
    while num2 != 0:
        num1, num2 = num2, num1 % num2
    return num1

# --- main execution ---
if __name__ == "__main__":
    try:
        #taking user input
        num1 = input("Enter the first number : ")
        num2 = input("Enter the second number : ")
        
        # if input is empty
        if not num1 or not num2:
            raise ValueError("Input can not be empty")

        # converting inside try block
        try:
            number1 = int(num1)
            number2 = int(num2)
        except ValueError:
            raise TypeError("Please valid integer only")
            
        #calling the function
        result = calculate_GCD(number1, number2)
        # print and display the result
        print(f"GCD is : {result}")
        
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)
```
