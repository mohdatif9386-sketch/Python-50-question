```python
def is_all_digit(text: str) -> bool:
    """
    function to check if a string contains only digit or not.
    args:
        (text: str) - input string.
    return:
        bool: (True) if string is digit otherwise False.
    raise: 
        TypeError: if input is not a string.
        ValueError: if input is empty.
    """
    # validate that not input
    if not isinstance(text, str):
        raise TypeError("input must be a string")
    # validate that input i empty
    if not text:
        ValueError("Input acnnot be empty")

    # check string
    for char in text:
        if char < "0" or char > "9":
            return False
    return True

# -- main execution --
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter the string to check :")
        # calling th function
        result = is_all_digit(user_input)
        print(result)
    except Exception as e:
        print("Error :", e)
```
