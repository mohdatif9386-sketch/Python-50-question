```python
def find_lcm(num1: int, num2: int) -> int:
    """
    function to find lcm of two number.
    args:
        (num1: int, num2: int) -- n=input num1 and num2
    return:
        int: lcm of num1 and num2
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(num1, int) and not isinstance(num2, int):
        raise TypeError("Input must be an integer")
    # negative check
    if num1 < 0 and num2 < 0:
        raise ValueError("input must be a positive number")

    # logic
    return abs(num1 * num2) // math.gcd(num1, num2)

# -- main execution --
if __name__ == "__main__":
    try:
        # taking inpput from the user
        num1 = input("Enter number 1 :")
        num2 = input("Enter number 2 :")

        # validate that not input
        if not num1 and not num2:
            raise ValueError("Input cannot be empty")

        # converting inside try block
        try:
            number1 = int(num1)
            number2 = int(num2)
        except ValueError:
            raise TypeError("Input must be an integer")
        # calling the function
        result = find_lcm(number1, number2) 
        # print and display the result
        print("LCM is :",result)

    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
