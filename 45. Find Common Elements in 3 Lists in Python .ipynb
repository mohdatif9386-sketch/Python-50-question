```python
list1 = [1, 2, 3, 4, 5, 6, 8]
list2 = [4, 5, 6, 7, 8]
list3 = [5, 6, 7, 8, 9]
```

```python
common_element = []
for item in list1:
    if item in list2 and item in list3 and item not in common_element:
        common_element.append(item)
print(common_element)
```

```python
def find_common_element(list1: list, list2: list, list3: list) -> list:
    """
    function to find common element in three list using for loop.
    args:
        input all three list.
    return:
         return a list of common element.
    raise:
        TypeError: if input is not a list.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(list1, list) or not isinstance(list2, list) or not isinstance(list3, list):
        raise TypeError("input must be a list")
    # validate that input is empty
    if not list1 or not list2 or not list3:
        raise ValueError("input cannot be empty")

    # logic
    common = []
    for item in list1:
        if item in list2 and item in list3 and item not in common:
            common.append(item)
    return common

# -- aain execution --
if __name__ == "__main__":
    try:
        # TAKING INPUT FROM THE USER
        list1 = [1, 2, 3, 4, 5, 6, 8, 9]
        list2 = [4, 5, 6, 7, 8, 9]
        list3 = [5, 6, 7, 8, 9]

        # calling the function
        result = find_common_element(list1, list2, list3)
        print("Commnon Element :", result)

    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
