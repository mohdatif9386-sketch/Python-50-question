```python
def count_character(input_text: str) -> dict:
    """
    function to count the occurance of each character using for loop.

    Args:
        input_text: str - The string whose character are to be count.
        
    return:
        dict: Given a dictionary with character as key and as count value.

    raise:
        TypeError: if input is not a string.
    """
    # validate that input type
    if not isinstance(input_text, str):
        raise TypeError("Input must be a string")
    # if input is empty
    if not input_text:
        raise ValueError("Input can not be empty")

    # logic
    char_count = {}
    for char in input_text.lower().replace(" ", ""):
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

# --- main execution ---
if __name__ == "__main__":
    try:
        #taking input from the user
        user_input = input("Enter the string to count occurrence of each character : ")
        # calling he function
        result = count_character(user_input)
        #print and display the result
        print(result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print(e)
```
