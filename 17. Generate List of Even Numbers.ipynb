```python
def list_even_number(number: int) -> int:
    """
    A function that generates an even-numbered list.
    Args:
        (number: int) - Input number
    Return:
        list: An even numbers given in the list.
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is empty.
    """
    # validate that inpt type
    if not isinstance(number, int):
        raise TypeError("Input must be an integer")
    
    # if input number is negative
    if number < 0:
        raise ValueError("Input must be a positive number")
        
    # logic
    even_list = []
    for i in range(2, number + 1):
        if i % 2 == 0:
            even_list.append(i)
    return even_list

# --- main execution ---
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter the number : ")
        # if input is empty
        if not user_input:
            raise ValueError("input canot be empty")
            
        # converting inside try block
        try:
            numbers = int(user_input)
        except ValueError:
            raise TypeError("User input give must be an integer value")
            
        # calling the function
        result = list_even_number(numbers)
        
    except (TypeError, ValueError) as error:
        print("Error :", error)
    except Exception as error:
        print("Unexpected Error : ", error)
        
    # print and display the result
    else:
        print(result)
```
