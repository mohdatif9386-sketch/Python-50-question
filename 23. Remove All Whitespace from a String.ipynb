```python
def remove_whitespace(text: str) -> str:
    """
    function to remove all whitespace characters from a string.

    Args:
        text (str): Input string

    Returns:
        str: String without whitespace

    Raises:
        TypeError: if input is not a string.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(text, str):
        raise TypeError("input must be a string")
    if not text:
        raise ValueError("Input cannot be empty")

    #logic
    result = "".join(text.split())
    return result

# --- main execution---
if __name__ == "__main__":
    try:
        # Taking input from the user
        user_input = input("Enter text to remove all whitespace : ")
        
        # calling the function
        result = remove_whitespace(user_input)
        
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)

    else:
        print("After removing all whitespace :", result)
```
