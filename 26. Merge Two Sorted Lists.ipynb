```python
def merge_two_list(list1: list, list2: list) -> list:
    """
    Merge two sorted lists into one sorted list.

    Args:
        list1 (list): First sorted list
        list2 (list): Second sorted list

    Returns:
        list: Merged sorted list

    Raises:
        TypeError: if input is not a list
        ValueError: if input is empty.
    """
    # vlidate that input type
    if not isinstance(list1, list) or not isinstance(list2, list):
        raise TypeError("both input must be a list")
    # if input is empty
    if not list1 or not list2:
        raise ValueError("Input cannot be empty")

    # logic
    merge = sorted(list1 + list2)
    return merge
    
# --- main execution ---
if __name__ == "__main__":
    #taking input from the user
    list1 = [1, 2, 3]
    list2 = [3, 4, 5]
    try:
        #clling the function
        result = merge_two_list(list1, list2)
        print(result)
    except TypeError as e:
        print("Type Error : ", e)
    except ValueError as e:
        print("Value Error : ", e)
    except Exception as e:
        print("Unexpected Error : ", e)
```
