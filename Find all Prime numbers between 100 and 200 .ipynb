```python
num = 102
for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
        print("not a prime number")
        break
else:
    print("prime number")
```

```python
def prime_number(number: int) -> bool:
    """
    function to check whether a number is prime or not
    Args:
        number: int - The number to check
    return:
        bool: True - if the number is prime and False otherwise
    raise:
        TypeError: if the input is not integer
        ValueError: if the number is less than 2
    """
    # validate that input type
    if not isinstance(number, int):
        raise TypeError("input must be a integer")
    # validate number at least 2
    if number < 2:
        raise ValueError("prime number is greater than 2 or equal")

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    else:
        return True

# -- main execution--
if __name__=="__main__":
    try:
        # taking input lower and upper limit from the user
        lower_limit = input("Enter lower limit : ")
        upper_limit = input("Enter upper limit : ")

        # validate and convert input into integer
        if not lower_limit.strip().isdigit():
            raise TypeError("Lower limit must be a valid integer (no letter, no string, no decimal")
        if not upper_limit.strip().isdigit():
            raise TypeError("Upper limit must be a valid integer (no letter, no string, no decimal")

        lower_limit = int(lower_limit)
        upper_limit = int(upper_limit)

        for i in range(lower_limit,upper_limit):
            if prime_number(i):
                print(f'{i} is a prime number')
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
