```python
def reverse_str(input_str: str) -> str:
    """
    function to reverse a given string using a for loop.
    Args:
        input_str: the string to be reverse.
    reverse:
        the reversed version of string.
    raises:
        TypeError: if input is not a string
    """
    # validate the input type
    if not isinstance(input_str, str):
        raise TypeError("Input must be string")
    reversed_str = ""
    for i in input_str:
        reversed_str = i + reversed_str
    return reversed_str

# --- main execution-- 
if __name__ == "__main__":
    try:
        # taking input from the user
        user_input = input("Enter your string to reversed : ")
        # calling the function
        result = reverse_str(user_input)
        # display the reversed strng
        print(f"reversed string : {result}")
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
