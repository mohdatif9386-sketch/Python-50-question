```python
def swap_number(num1: int, num2: int) -> int:
    """
    function to swap number without temporary variables.
    args:
        (num1: int, num2: int) - Input to swap number.
    return:
        int: After swap number
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(num1, int) and not isinstance(num2, int):
        raise TypeError("Input must be an integer")
    # validate that input is empty
    if not num1 and not num2:
        raise ValueError("Input cannot be empty")

    # logic
    num1, num2 = num2, num1
    return num1, num2

# -- main execution --
if __name__ == "__main__":
    try:
        # taking iinput from the user
        input_num1 = input("Enter number 1 :")
        input_num2 = input("Enter number 2 :")

        # validate that input is empty
        if not input_num1 and not input_num2:
            raise ValueError("Input cannot be empty")

        # converting inside try block
        try:
            num1 = int(input_num1)
            num2 = int(input_num2)
        except ValueError:
            raise TypeError("input must be an integer")
        
        # calling the function
        num1, num2 = swap_number(num1, num2)
        # print and display the result
        print("num1 after swap :", num1)
        print("num2 after swap :", num2)

    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
