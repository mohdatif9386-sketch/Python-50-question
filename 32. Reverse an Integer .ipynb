```python
def reverse_integer(number: int) -> int:
    """
    function to reverse a number using while loop.
    args:
        (number: int) -- input number to reverse.
    returns:
        inr: will reverse the given number
    raise:
        TypeError: if input is not an integer.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(number, int):
        raise TypeError("Input must be an integer")

    # logic
    sign = -1 if number < 0 else 1
    number = abs(number)
    reverse_int = 0
    while number > 0:
        digit = number % 10
        # print(digit)
        reverse_int =reverse_int * 10 + digit
        number = number // 10
    return sign * reverse_int

# -- main execution ---
if __name__ == "__main__":
    try:
        # taking inout from the user
        user_input = input("Enter the number to reverse :")
        # validate that not input
        if not user_input:
            raise ValueError("Input cannot be empty")
        # converting inside try bloack
        try:
            numbers = int(user_input)
        except ValueError:
            raise TypeError("Input must be an integer")

        # calling thr function
        result = reverse_integer(numbers)
        #print and display the result
        print("After reverse integer :", result)

    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exception as e:
        print("Error :", e)
```
