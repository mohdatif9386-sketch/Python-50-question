```python
def factorial(number: int) -> int:
    """
    function to calculate the factorial of a non-negaive integer
    Args:
        number: int - The number for which factorial is required.
    Return:
        int: factorial of the given number.
    raise:
        TypeError: if the number is negative.
        ValueError: if the number is non - integer
    """

    # validate that type input
    if not isinstance(number, int):
        raise TypeError("input must be a numeric value")
    if isinstance(number, bool):
        raise TypeError("boolean value are not accepted")
    # negative check
    if number < 0:
        raise ValueError("negative number does not exist")
    
    factorial = 1
    for i in range(1, number+1):
        factorial = factorial * i
    return factorial
    
# --main execution--
if __name__=="__main__":
    try:
        # taking the input from the user
        user_input = input("Enter the number to find factorial : ")
        # convert the input
        numbers = int(user_input)
        # calling the function
        result = factorial(numbers)
        # display the result
        print(f"factorial of {numbers} is : {result}")
    except (TypeError, ValueError) as error:
        print("input must be a number")
    except Exception as e:
        print(e)
```
