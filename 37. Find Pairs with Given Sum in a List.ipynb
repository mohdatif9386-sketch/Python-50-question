```python
nums = [1, 2, 3, 4, 5, 6]
target = 7
pairs = []
for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == target:
            pairs.append((nums[i], nums[j]))
print(pairs)
```

```python
def pairs_nums(List: list, target: int) -> list:
    """
    function to find pairs with given sum in a list using for loop.
    args:
        (List: list, target: int) - input list in an integer and target.
    return:
        list: given a number of number equal of target.
    raise:
        TypeError: if input is not a list.
        ValueError: if input is empty.
    """
    # validate that input type
    if not isinstance(List, list):
        raise TypeError("Input must be a list in a integer")
    #validate that target type
    if not isinstance(target, int):
        raise TypeError("input must be an integer")
    # check lenght
    if len(List) == 0:
        raise ValueError("Input cannot be empty")
    # list len is greater than two or more
    if len(List) == 1:
        raise ValueError("list has contain two or more elements")

    # logic
    pair = []
    for i in range(len(List)):
        for j in range(i+1, len(List)):
            if List[i] + List[j] == target:
                pair.append((List[i], List[j]))
    return pair

# -- main execution --
if __name__ == "__main__":
    # taking input from the user
    user_input = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    target = 10
    try:
        #calling the function
        result = pairs_nums(user_input, target)
        # print and display the result
        print(result)
    except (TypeError, ValueError) as e:
        print("Error :", e)
    except Exceptionc as e:
        print("Error :", e)
```
